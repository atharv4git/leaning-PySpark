```python
data.printSchema()
```

```python
# imputing
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=['age','score'],
    outputCols=[f'{c}_imputed' for c in ['age','score']]
).setStrategy("mean")
```

```python
data_imputed = imputer.fit(data).transform(data)
```

```python
data_imputed.show()
```

```python
data_imputed.filter("age_imputed>=21").show()
```

```python
data_imputed.filter("age_imputed>=21").select("Name").show()
```

```python
# using MLib
data = spark.read.csv("test4.csv",header=True, inferSchema=True)
data.show()
```

```python
data.printSchema()
```

```python
data.columns
```

```python
from pyspark.ml.feature import VectorAssembler
vec_ass = VectorAssembler(inputCols=['age','experience'],outputCol="IndependentFeatures")
```

```python
output = vec_ass.transform(data)
output.show()
```

```python
output.columns
```

```python
finalized_data = output.select("IndependentFeatures","salary")
```

```python
finalized_data.show()
```

```python
from pyspark.ml.regression import LinearRegression

train_data, test_data = finalized_data.randomSplit([0.75,0.25])

regressor = LinearRegression(featuresCol='IndependentFeatures', labelCol='salary')

regressor = regressor.fit(train_data)
```

```python
regressor.coefficients
```

```python
regressor.intercept
```

```python
pred_lr = regressor.evaluate(test_data)
```

```python
pred_lr.predictions.show()
```
